```python
import numpy as np
import pandas as pd
```

```python
inbound = pd.read_csv('../Data/Inbound.csv')
outbound = pd.read_csv('../Data/Outbound.csv')
inventory = pd.read_csv('../Data/Inventory.csv')
operation = pd.read_csv('../Data/OperationCost.csv')
material = pd.read_csv('../Data/MaterialMaster.csv')
```

```python
inventory
```

```python
scrap_rows = inventory[inventory['BATCH_NUMBER'] == 'SCRAP']
scrap_rows
```

```python
# Inventory + Material Master
df_inventory_material = pd.merge(inventory, material, on='MATERIAL_NAME', how='left') 
df_inventory_material
```

```python
# Total of scrap batches in inventory
scrap_batches = df_inventory_material[df_inventory_material['BATCH_NUMBER'] == 'SCRAP']
print(f"Total scrap batches: {len(scrap_batches)}")
```

```python
# Unique batches per material 
grouped = df_inventory_material.groupby('MATERIAL_NAME')['BATCH_NUMBER'].unique().reset_index()

grouped['NUM_BATCHES'] = grouped['BATCH_NUMBER'].apply(len)

grouped
```

```python
print("Inbound =: \n", inbound.isnull().sum())
print("\n")
print("Inventory =: \n", inventory.isnull().sum())
print("\n")
print("Material =: \n", material.isnull().sum())
print("\n")
print("Operation =: \n", operation.isnull().sum())
print("\n")
print("Outbound =: \n", outbound.isnull().sum())
```

```python
inventory_cleaned = inventory.dropna(subset=['BATCH_NUMBER'])
material_cleaned = material.dropna(subset=['MATERIAL_NAME'])

print("Inventory =: \n", inventory_cleaned.isnull().sum())
print("Material =: \n", material_cleaned.isnull().sum())
```

```python
# group in bound as per date, plant, and material
in_group = inbound.groupby(['INBOUND_DATE', 'PLANT_NAME', 'MATERIAL_NAME']).agg(
    total_quantity_mt=pd.NamedAgg(column='NET_QUANTITY_MT', aggfunc='sum')
).reset_index()

in_group
```

```python
# group out bound as per date, plant, and material
out_group = outbound.groupby(['OUTBOUND_DATE','PLANT_NAME', 'MATERIAL_NAME']).agg(
    total_quantity_mt=pd.NamedAgg('NET_QUANTITY_MT', 'sum')
).sort_values(by='OUTBOUND_DATE').reset_index()

out_group
```

```python
initial_inv = inventory.groupby(['BALANCE_AS_OF_DATE', 'PLANT_NAME', 'MATERIAL_NAME']).agg(
    InitialInv=('UNRESRICTED_STOCK', 'sum')
).reset_index()

# Convert the column to datetime format
initial_inv['BALANCE_AS_OF_DATE'] = pd.to_datetime(initial_inv['BALANCE_AS_OF_DATE'], format='%m/%d/%Y')

initial_inv[initial_inv['MATERIAL_NAME'] == 'MAT-0050'].sort_values(by='BALANCE_AS_OF_DATE', ascending=True)
```

```python
initial_div_1000 = initial_inv['InitialInv']/1000
initial_inv['InitialInv'] = initial_div_1000
initial_inv[initial_inv['MATERIAL_NAME'] == 'MAT-0050'].sort_values(by='BALANCE_AS_OF_DATE', ascending=True)

```

```python
# inventory in ascending order of date

inventory['BALANCE_AS_OF_DATE'] = pd.to_datetime(inventory['BALANCE_AS_OF_DATE'])
inventory.sort_values(by='BALANCE_AS_OF_DATE', ascending=True)
inventory 
```

```python
# inbound in ascending order of date
inbound['INBOUND_DATE'] = pd.to_datetime(inbound['INBOUND_DATE'])
inbound.sort_values(by='INBOUND_DATE', ascending=True)
inbound 
```

```python
# outbound in ascending order of date
outbound['OUTBOUND_DATE'] = pd.to_datetime(outbound['OUTBOUND_DATE'])
outbound.sort_values(by='OUTBOUND_DATE', ascending=True)
outbound
```

```python
# inventory for china plant
china_inventory = inventory[inventory['PLANT_NAME'] == 'CHINA-WAREHOUSE']
china_inventory
```

```python
# sum unrestricted stock for china plant
initial_inv_china = china_inventory.groupby(['BALANCE_AS_OF_DATE', 'MATERIAL_NAME']).agg(
    InitialInv=('UNRESRICTED_STOCK', 'sum')
).reset_index()

initial_inv_china

```

```python
# turn stock from KG to MT china
initial_inv_china_1000 = initial_inv_china['InitialInv']/1000
initial_inv_china['InitialInv'] = initial_inv_china_1000
initial_inv_china
```

```python
min_date = initial_inv_china['BALANCE_AS_OF_DATE'].min()
max_date = pd.to_datetime("2024-12-31")
full_dates = pd.date_range(min_date, max_date)
```
