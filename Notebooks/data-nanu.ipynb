```python
# get outbound for china plant per date and material
china_outbound = outbound[outbound['PLANT_NAME'] == 'CHINA-WAREHOUSE'].copy()
outbound_grouped = china_outbound.groupby(['OUTBOUND_DATE', 'MATERIAL_NAME'])['NET_QUANTITY_MT'].sum().reset_index()
outbound_grouped.rename(columns={'OUTBOUND_DATE': 'Day', 'NET_QUANTITY_MT': 'OutboundQty'}, inplace=True)
outbound_grouped[outbound_grouped['MATERIAL_NAME'] == 'MAT-0001'].sort_values(by='Day', ascending=True)
```

```python
# inventory for china plant
china_inventory = inventory[inventory['PLANT_NAME'] == 'CHINA-WAREHOUSE']
china_inventory
```

```python
# sum unrestricted stock for china plant
initial_inv_china = china_inventory.groupby(['BALANCE_AS_OF_DATE', 'MATERIAL_NAME']).agg(
    InitialInv=('UNRESRICTED_STOCK', 'sum')
).reset_index()

initial_inv_china

```

```python
# turn stock from KG to MT china
initial_inv_china_1000 = initial_inv_china['InitialInv']/1000
initial_inv_china['InitialInv'] = initial_inv_china_1000
initial_inv_china
```

```python
# list all materials in china plant
china_materials = initial_inv_china['MATERIAL_NAME'].unique()
print("Materials in China Plant:", china_materials)
```

```python
min_date_china = initial_inv_china['BALANCE_AS_OF_DATE'].min()
max_date_china = pd.to_datetime("2024-12-31")
full_dates = pd.date_range(min_date_china, max_date_china)

# Cross join dates with material list
materials = initial_inv_china['MATERIAL_NAME'].unique()
date_df = pd.DataFrame({'Day': full_dates})
mat_df = pd.DataFrame({'MATERIAL_NAME': materials})
cross = date_df.merge(mat_df, how='cross')

# Merge and forward fill
daily_inv = cross.merge(
    initial_inv_china,
    left_on=['Day', 'MATERIAL_NAME'],
    right_on=['BALANCE_AS_OF_DATE', 'MATERIAL_NAME'],
    how='left'
)
daily_inv = daily_inv.sort_values(['MATERIAL_NAME', 'Day'])
daily_inv['InitialInv'] = daily_inv.groupby('MATERIAL_NAME')['InitialInv'].ffill()

daily_inv['InitialInv'] = daily_inv['InitialInv'].fillna(0)

# Clean up
daily_inv = daily_inv[['Day', 'MATERIAL_NAME', 'InitialInv']]
daily_inv
```

```python
daily_inv = daily_inv.merge(outbound_grouped, on=['Day', 'MATERIAL_NAME'], how='left')
daily_inv['OutboundQty'] = daily_inv['OutboundQty'].fillna(0)
```

```python
def propagate_inventory(group):
    group = group.sort_values('Day').copy()
    corrected = []
    current = None
    for _, row in group.iterrows():
        if current is None:
            current = row['InitialInv'] + row['OutboundQty']
        else:
            current += row['OutboundQty']
        corrected.append(current)
    group['PropagatedInv'] = corrected
    return group

# Apply per MATERIAL_NAME
daily_inv = daily_inv.sort_values(['MATERIAL_NAME', 'Day'])
daily_inv = daily_inv.groupby('MATERIAL_NAME', group_keys=False).apply(propagate_inventory)

daily_inv[daily_inv['MATERIAL_NAME'] == 'MAT-0430'].sort_values(by='Day', ascending=True)
```

```python
# inventory for singapore plant
singapore_inventory = inventory[inventory['PLANT_NAME'] == 'SINGAPORE-WAREHOUSE']
singapore_inventory
```

```python
# sum unrestricted stock for singapore plant
initial_inv_singapore = singapore_inventory.groupby(['BALANCE_AS_OF_DATE', 'MATERIAL_NAME']).agg(
    InitialInv=('UNRESRICTED_STOCK', 'sum')
).reset_index()
initial_inv_singapore
```

```python
# turn stock from KT to MT singapore

initial_inv_singapore_1000 = initial_inv_singapore['InitialInv']/1000
initial_inv_singapore['InitialInv'] = initial_inv_singapore_1000
initial_inv_singapore
```

```python
min_date_singapore = initial_inv_singapore['BALANCE_AS_OF_DATE'].min()
max_date_singapore = pd.to_datetime("2024-12-31")
full_dates_singapore = pd.date_range(min_date_singapore, max_date_singapore)

# Cross join dates with material list
materials_singapore = initial_inv_singapore['MATERIAL_NAME'].unique()
date_df_singapore = pd.DataFrame({'Day': full_dates_singapore})
mat_df_singapore = pd.DataFrame({'MATERIAL_NAME': materials_singapore})
cross_singapore = date_df_singapore.merge(mat_df_singapore, how='cross')

# Merge and forward fill
daily_inv_singapore = cross_singapore.merge(
    initial_inv_singapore,
    left_on=['Day', 'MATERIAL_NAME'],
    right_on=['BALANCE_AS_OF_DATE', 'MATERIAL_NAME'],
    how='left'
)
daily_inv_singapore = daily_inv_singapore.sort_values(['MATERIAL_NAME', 'Day'])
daily_inv_singapore['InitialInv'] = daily_inv_singapore.groupby('MATERIAL_NAME')['InitialInv'].ffill()

# Clean up
daily_inv_singapore = daily_inv_singapore[['Day', 'MATERIAL_NAME', 'InitialInv']]
daily_inv_singapore
```
